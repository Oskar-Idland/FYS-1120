## Krets drevet av en firkantpuls

Vi skal i denne oppgaven studere kretsen som er vist i figuren under. Kretsen består av en spenningskilde, $U(t)$, en kondensator $C$ og en motstand $R$.

### Oppgave (b)
Finn en differensiallikning for spenningen $V_C$ over kondensatoren for en vilkårlig spenningskilde $U(t)$.

(*Svar her*)

__Løsning__ Vi bruker Kirchoffs spenningslov
$$U(t) - RI - V_C = 0$$
Her er $V_C = Q/C$. Strømmen $I$ er gitt som $dQ/dt$ slik at $dV_c/dt = (1/C)dQ/dt = I/C$ slik at $I = C dV_C/dt$. Vi setter dette inn i likningen
$$U(t) - RC \frac{dV_C}{dt} - V_C = 0$$
Likningen for $V_C$ blir derfor 
$$\frac{d V_C}{dt} = \frac{1}{RC}\left( U(t) - V_C \right)$$

### Oppgave (c)
Finn $V_C(t)$ når spenningskilden er en firkantpuls:
$$U(t) = \text{sign}(\sin \omega t)$$
hvor $\text{sign}(x) = +1$ når $x>0$ og $-1$ når $x<0$. Ved tiden $t=0$ er $I=0$. Du kan dessuten bruke at $R = 100$Ohm, $C = 10^{-3}$F, og $\omega = 2 \pi /T$ hvor $T = 1.0$s.
Skriv et program som finner strømmen og visualiser denne for et passende tidsintervall.

```python
# Løsning
# Vi løser diff-likningen dV/dt = (U-V)/(RC)
# ved Eulers metode 
import numpy as np
import matplotlib.pyplot as plt
R = 100.0
C = 1.0e-3
RC = 0.1
T = 1.0
omega = 2*np.pi/T
dt = 0.0001
time = 10.0
n = int(time/dt)
V = np.zeros(n)
t = np.zeros(n)
for i in range(n-1):
    U = np.sign(np.sin(omega*t[i]))
    V[i+1] = V[i] + dt*(U-V[i])/RC
    t[i+1] = t[i] + dt
```

```python
plt.plot(t,V)
```
